```python
#grph

from Bio import SeqIO

sequences = []
ids = []
counter_s = 0
counter_t = 0

with open('/Users/nocide51/Projects_Coding/psc2k22/asignment5/sequences/rosalind_grph.txt', 'r') as handle:
    for x in SeqIO.parse(handle, "fasta"):
        sequences.append(x)
        ids.append(x.id)
    for seq in sequences:
        counter_t = 0
        s = list(seq[-3:])
        s_id = ids[counter_s]
        #print(s)
        for seq2 in sequences:
            t = list(seq2[:3])
            t_id = ids[counter_t]
            #print(t)
            if s == t:
                if s_id == t_id:
                    continue
                else:
                    print(s_id + ' ' + t_id)
                
            counter_t += 1
        counter_s += 1
```
