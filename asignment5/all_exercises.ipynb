```python
#splc
import Bio

def spliceosome(dna, intron):
    dna = str(dna)
    intron = str(intron)
    if intron in dna:
        start = dna.index(intron)
        length = len(intron)
        end = start+length
        first_part = dna[:start]
        intron_part = dna[start:end]
        second_part = dna[end:]
        without_intron = first_part + second_part
        #print(without_intron)
        return without_intron



sequences = []
with open('/Users/nocide51/Projects_Coding/psc2k22/asignment5/sequences/rosalind_splc.txt', 'r') as handle:
    for x in SeqIO.parse(handle, 'fasta'):
        sequences.append(x.seq)
    dna = sequences[0]
    print(dna)
    #spliceosome(dna, sequences[1])
    print(sequences[1:])
    new_dna = []
    for x in range(1, len(sequences)):
        if x == 1:
            new_dna.append(spliceosome(dna, sequences[x]))
        else:
           new_dna.append(spliceosome(new_dna[-1], sequences[x]))
    print(new_dna[-1])
    prot_seq = Seq(new_dna[-1]).translate()
    print(prot_seq)
        
```

```python
#pmch

from Bio import SeqIO
from math import factorial

def get_perfect(s):
    AU_matches = 0
    GC_matches = 0
    for x in s:
        if x == 'A':
            AU_matches += 1
        if x == 'G':
            GC_matches += 1
    result = factorial(AU_matches)*factorial(GC_matches)
    return result

with open('/Users/nocide51/Projects_Coding/psc2k22/asignment5/sequences/rosalind_pmch.txt', 'r') as handle:
    for x in SeqIO.parse(handle, 'fasta'):
        sequence = x.seq
    print(get_perfect(sequence))
```

```python
#long solution that doesn't work because it looks for the best match; which is not efficient

def consensus(seq):
    consensus=[]
    for seq1, seq2 in zip(seq[0][0],seq[0][1]):
        if seq1 == seq2:
            consensus.append(seq1)
        elif seq1 == '-':
            consensus.append(seq2)
        elif seq2 == '-':
            consensus.append(seq1)
    return ''.join(consensus)


from Bio import SeqIO
from Bio import pairwise2 
from Bio.pairwise2 import format_alignment 
from Bio.SubsMat import MatrixInfo as matlist
matrix = matlist.blosum62

reads = []


with open('/Users/nocide51/Projects_Coding/psc2k22/asignment5/sequences/rosalind_long.txt', 'r') as handle:
    for read in SeqIO.parse(handle, 'fasta'):
        reads.append(read.seq)
    #print(reads)
    for i in range(len(reads)):
        for x in range(len(reads)):
            alignment = pairwise2.align.localms(reads[i], reads[x], 2, -1, -3, -3)
            consensusseq = consensus(alignment)
        print(consensusseq)
```

```python
#long

from Bio import SeqIO
import io

#dynamic solution to #long, because a solution using biopython's pairwise doesn't work (too long, too ineficient for rosalind), the key is the 'half of the sequence is a match' which allow us to look only once for matching sequence, and not look for the best match possible.
def assembly(reads, superstring=''):
    #if genome has been assembled: return the assembly (contig)
    if len(reads) == 0:
        return superstring
    elif len(superstring) == 0: #first passage of the function, take the first seq from the reads to compare 
        superstring = reads.pop(0)
        #print(superstring)
        return assembly(reads, superstring)
    else:#normal parsing of the function until no reads is left to be read
        for i in range(len(reads)):
            current_read = reads[i]
            current_read_length = len(current_read)
            for x in range(current_read_length // 2): #get a 50% overlap, then keep passing the results through the function, elongating the superstring with the result
                overlap_length = current_read_length - x
                if superstring.startswith(current_read[x:]):
                    reads.pop(i)
                    return assembly(reads, current_read[:x] + superstring)
                if superstring.endswith(current_read[:overlap_length]):
                    reads.pop(i)
                    return assembly(reads, superstring + current_read[overlap_length:])

reads = []
with open('/Users/nocide51/Projects_Coding/psc2k22/asignment5/sequences/rosalind_long.txt', 'r') as handle:
    for record in SeqIO.parse(handle, 'fasta'):
        reads.append(str(record.seq))
    handle.close()
    result = assembly(reads)
    print(result)
```

```python
#tran

from Bio import SeqIO

data = []

with open('/Users/nocide51/Projects_Coding/psc2k22/asignment5/sequences/rosalind_tran.txt', 'r') as handle:
    for x in SeqIO.parse(handle, 'fasta'):
        x = str(x.seq)
        data.append(x)
    print(data)
    s1 = data[0]
    s2 = data[1]
    print(s1)
    transitions = 0
    transversions = 0
    for base in range(len(s1)):
        if s1[base] != s2[base]: #is there a mutation?
            if s1[base] == 'C' or s1[base] == 'T': #what was the original base?
                if s2[base] == 'C' or s2[base] == 'T': #what type of mutation is it
                    transitions += 1 #count it
                if s2[base] == 'A' or s2[base] == 'G':
                    transversions +=1
            if s1[base] == 'A' or s1[base] == 'G':
                if s2[base] == 'A' or s2[base] == 'G' :
                    transitions += 1 
                if s2[base] == 'C' or s2[base] == 'T':
                    transversions += 1
    print(transitions)
    print(transversions)
    print(transitions/transversions)
```
